# Analyzing	and	Visualizing	WeRateDogs

### INTRODUCTION
Avez-vous entendu parler de l'un des mèmes les plus populaires de 2016, "they're good dogs Brent", et du compte Twitter WeRateDogs ? Probablement oui, mais sinon.
Je vais vous en parler un peu ici, puis vous montrer une analyse amusante des données de tweet de WeRateDogs.
Pour commencer, WeRateDogs est un compte Twitter qui évalue les chiens des gens et inclut un commentaire humoristique sur le chien. Ces notes ont presque toujours un dénominateur de 10. Les numérateurs, en revanche, sont presque toujours supérieurs à 10 (11/10, 12/10, 13/10, etc.). En théorie, les taux devraient aller de 1 à 10. WeRateDogs, d'autre part, estime que presque tous les chiens méritent un 10 et parfois plus. WeRateDogs compte plus de 6 millions de followers et a été présenté dans les médias internationaux. L'une de ces histoires concernait la citation "they're good dogs Brent", qui était un échange dans lequel WeRateDogs a fermé une personne qui avait des problèmes avec son système de notation de manière amusante.
